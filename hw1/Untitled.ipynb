```python
with open('testcase/input1.txt','r') as file:
    file = file.readlines()
```

```python
int(file[0].split(' ')[1])
```

```python
edge = int(file[0].split(' ')[1])
node = int(file[0].split(' ')[0])
```

# generate link list

```python
graph = [([]) for i in range(node+1)]
in_degrees = [0 for x in range(node+1)]
for i in range(edge):
    n = int(file[i].split(' ')[0])
    connect_node = int(file[i+1].split(' ')[1])
    in_degrees[connect_node] += 1
#     print(connect_node)
    graph[n].append(connect_node)
    
```

```python
temp = [x for x in range(node)]
graph_dict = dict(zip(temp,graph))
```
